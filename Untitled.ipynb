```python
import pathlib
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image 
%matplotlib inline
```

```python
path = pathlib.Path.cwd()/"Romeo and Juliet.txt"
print(path)
```

```python
text = path.read_text()
wordcloud = WordCloud().generate(text)

plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig("Romeo and Juliet.png", dpi = 600, facecolor = "w", format= "png")
plt.show()
```
